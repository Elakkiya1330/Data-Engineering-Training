#1. Create a CSV of Sales Data
* Save this as sales_data.csv in Colab.
* Read it using pd.read_csv() and print the full DataFrame.

```python
import pandas as pd

csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000

Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""
with open('sales_data.csv', 'w') as file:
  file.write(csv_data)

df = pd.read_csv('sales_data.csv')
print(df)
```

#2. Total Sales Per Store
* Group the data by Store and calculate total sales for each store.
* Identify which store had higher total sales.

```python
total_sales = df.groupby('Store')['Sales'].sum()
print(total_sales)
```

```python
higher = total_sales[total_sales == total_sales.max()].index[0]
print("Store with higher total sales: ",higher)
```

#3. Monthly Sales Trend
* Group the data by Month and sum up the total sales (Store A + Store B).
Print the result.

```python
monthly = df.groupby('Month')['Sales'].sum()
print(monthly)
```

#4. Chart: Monthly Sales by Store
* Create a line chart showing monthly sales for both Store A and Store B.
  
    X-axis: Month

    Y-axis: Sales

    Two lines: Store A and Store B

```python
import matplotlib.pyplot as plt
pivot_df = df.pivot(index="Month", columns="Store", values="Sales").plot(marker ='o')
plt.title("Subject Scores by Student")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#5. Bonus
* Create a bar chart showing total sales per store.
* Add data labels on top of each bar.
* Save this chart as an image ( sales_chart.png ).

```python
stores = total_sales.index.tolist()
sales = total_sales.values.tolist()
plt.bar(stores, sales, color=['skyblue', 'lightgreen'])
for i in range(len(stores)):
    plt.text(i, sales[i] + 500, str(sales[i]), ha='center')
plt.title("Total Sales per Store")
plt.ylabel("Sales")
plt.show()
```
