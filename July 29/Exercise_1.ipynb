#1. Create a CSV and Read It

```python
import pandas as pd
```

```python
# Read it into a pandas DataFrame and print the first 3 rows.
df = pd.read_csv('/content/drive/My Drive/students.csv')
print(df.head(3))
```

# 2. Calculate Average Scores
* Add a new column Average that calculates the mean score for Math, Science, and English for each student
* Print the student with the highest average.




```python
df['Average'] = round(df[['Math','Science','English']].mean(axis=1),2)
print(df)
```

```python
highest_avg = df[df['Average'] == df['Average'].max()]
print("The student with highest average is :\n\n",highest_avg)
```

#3. Filter Students
* Display all students who have scored more than 80 in Science.
* Display all students with an average score below 70.

```python
print("students scored more than 80:\n\n",df[df['Science'] > 80])
```

```python
print("students with average score less than 70:\n\n",df[df['Average']<70])
```

#4. Sorting
* Sort the students by English score (descending) and print the top 3.

```python
desc_eng = df.sort_values(by='English', ascending=False)
print(desc_eng.head(3))
```

#5. Chart Exercise
* Create a bar chart showing students’ names on the X-axis and their average
scores on the Y-axis.
* Create a line chart for all students’ scores in Math, Science, and English (use df.plot() ).

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['Name'],df['Average'],color = 'yellow')
plt.xlabel('Name')
plt.ylabel('Average_scores')
plt.title('Students_average')
plt.xticks(rotation = 25)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.set_index('Name')[['Math','Science','English']].plot(kind='line',marker='o',figsize=(10,6))
plt.title("Subject scores by students")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#6. Bonus Task
* Find the subject with the highest total score across all students.
* Save the updated DataFrame (with Average column) as a new CSV named
students_summary.csv .

```python
subject_totals = df[['Math', 'Science', 'English']].sum()
highest = subject_totals[subject_totals == subject_totals.max()].index[0]
print("Subject with highest total score:", highest)
```
